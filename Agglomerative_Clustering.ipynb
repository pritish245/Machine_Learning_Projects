```python
data=df.drop("model",axis=1) #this data is needed to create dendogram and for modeling part
models=df["model"]  #this is for reference for the label
```

```python
data
```

```python
models
```

```python
from scipy.cluster.hierarchy import dendrogram, linkage
```

```python
Z=linkage(data,method="ward",metric="euclidean")
```

```python
# Columns of linkage matrix = 1st cluster, 2nd cluster, distance, no. of samples in each cluster
```

```python
dendrogram(Z,labels=np.array(models))
plt.title("Dendrogram")
plt.xlabel("Models")
plt.xticks(rotation=90)
plt.ylabel("Distance")
plt.show()
```

```python
from sklearn.cluster import AgglomerativeClustering
```

```python
model=AgglomerativeClustering(n_clusters=2,affinity="euclidean",linkage="ward")
```

```python
model.fit(data)
```

```python
df["ClusterID"]=model.labels_
```
