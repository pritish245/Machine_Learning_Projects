```python
unemployment_correlation = data_new.groupby('Store')[['Weekly_Sales', 'Unemployment']].corr().unstack().iloc[:, 1]
print("\nCorrelation between Weekly Sales and Unemployment for each store:\n", unemployment_correlation)
```

```python
most_affected_stores = unemployment_correlation.sort_values().head(5)
print("\nStores most affected by Unemployment:\n", most_affected_stores)
```

```python
holiday_impact = data_new.groupby('Holiday_Flag')['Weekly_Sales'].mean()
print("\nImpact of Holiday Flag on Weekly Sales:\n", holiday_impact)
```

```python
temperature_correlation = data_new['Weekly_Sales'].corr(data_new['Temperature'])
print("\nCorrelation between Weekly Sales and Temperature:", temperature_correlation)
```

```python
cpi_correlation = data_new['Weekly_Sales'].corr(data_new['CPI'])
print("\nCorrelation between Weekly Sales and CPI:", cpi_correlation)
```

```python
total_sales_per_store = data_new.groupby('Store')['Weekly_Sales'].sum()
top_stores = total_sales_per_store.sort_values(ascending=False).head(5)
print("\nTop Performing Stores:\n", top_stores)
```

```python
worst_store = total_sales_per_store.sort_values().head(1)
print("\nWorst Performing Store:\n", worst_store)

highest_sales = total_sales_per_store.max()
lowest_sales = total_sales_per_store.min()
sales_difference = highest_sales - lowest_sales
percentage_difference = (sales_difference / highest_sales) * 100

print("\nDifference between highest and lowest performing stores: ", sales_difference)
print("Percentage Difference: ", percentage_difference, "%")
```

```python
# before building model we should check that the data is stationary or not.if the data is non-stationary then we convert to stationary model.we can identify he non stationarity of data by using augmented dicky fuller test
```

```python
# if p value <0.05 then that is stationary other wise non-stationary.
# we can make data stationary by taking log of data or differentiating the data
```

```python
# acf(auto correlation function) and pacf(partial auto correlation function) are used to find the order of model in arima and sarima
# arima-auto regressive integrated moving average
#sarima-seasonal auto regressive integrated moving average
```

```python
from statsmodels.tsa.stattools import acf,pacf
from statsmodels.graphics.tsaplots import plot_acf,plot_pacf
```

```python
acf_data=acf(final_data1)
pacf_data=pacf(final_data1)
```

```python
plot_acf(acf_data)
```

```python
plot_pacf(pacf_data,lags=10)
```

```python
# apply arima model
```

```python
from statsmodels.tsa.arima.model import ARIMA
```

```python
final_data1.shape
```

```python
train_data=final_data1.iloc[0:130]['Weekly_Sales']
test_data=final_data1.iloc[130:]['Weekly_Sales']
```

```python
final_data1[15:45]
```

```python
arima=ARIMA(train_data,order=(1,0,1))
trained_model=arima.fit()
```

```python
final_data1['Predicted']=trained_model.predict(start=len(train_data),end=len(final_data1)-1,dynamic=True)
```

```python
final_data1.tail(15)
```

```python
final_data1.plot()
```

```python
# apply sarima model
```

```python
from statsmodels.tsa.statespace.sarimax import SARIMAX
```

```python
sarima=SARIMAX(train_data,order=(1,0,1),seasonal_order=(1,0,1,12))
```

```python
sarima_model=sarima.fit()
```

```python
final_data1['Predicted']=sarima_model.predict(start=len(train_data),end=len(final_data1)-1,dynamic=True)
```

```python
final_data1.tail(15)
```

```python
final_data1.plot()
```

```python
forecasted_result=sarima_model.forecast(steps=12)
```

```python
#final_data1.plot()
forecasted_result.plot()
```
